```python
# import library
import pandas as pd
import geopandas as gpd
```

```python
# read the shape file
# sf stands for shape file
sf = gpd.read_file("../data/raw/PTV METRO TRAIN STATION/PTV_METRO_TRAIN_STATION.shp")

sf
```

```python
# Convert the geometry shaape to to latitude and longitude
# this code is taken from Applied Data Science workshop 
sf['geometry'] = sf['geometry'].to_crs("+proj=longlat +ellps=WGS84 +datum=WGS84 +no_defs")
sf
```

```python
csv_file_path = "../data/landing/train_station.csv"

# Convert the GeoDataFrame to CSV
sf.to_csv(csv_file_path, index=False)
```

```python
# create a JSON 
geoJSON = sf[['STOP_ID', 'geometry']].drop_duplicates('STOP_ID').to_json()

# print the first 300 chars of the json
print(geoJSON[:300])
```

```python
# create the map of the train station
import folium
_map = folium.Map(location=[-37.8136, 144.9631], tiles="Stamen Terrain", zoom_start=10)

# refer to the folium documentations on how to plot aggregated data.
_map.add_child(folium.Choropleth(
    geo_data=geoJSON,
    name='choropleth',
))

_map
```
