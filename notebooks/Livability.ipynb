```python
# Importing Libraries
import pandas as pd
```

```python
# Read SA2 Data
df = pd.read_csv("../Data/2021ABS_csv.txt", delimiter=",", index_col="SA2 Area Name")
df.head()
```

```python
# New feature
victoria_average_income = 1759
df['rental affordability'] = df['Median weekly household rental payment ($)'] / victoria_average_income
```

```python
# Load 2023 crime data
crime_df = pd.read_csv("../Data/recorded crime rates.csv")
crime_df = crime_df[crime_df['Year'] == 2023]
```

```python
crime_df.set_index("Local Government Area", inplace=True)
crime_df = crime_df[["Rate per 100,000 population"]]
crime_df = crime_df[crime_df.index != 'Total']
crime_df
```

```python
df['Cleaned_SA2_Name'] = df.index.str.replace('Region summary_ ', '').str.split(' SA2').str[0]
```

```python
link_regions = pd.read_csv('../Data/SA2LGA.csv')
```

```python
crime_df
```

```python
# Step 1: Create a mapping from Cleaned_SA2_Name to LGA
sa2_to_lga = link_regions.set_index('SA2')['LGA'].to_dict()
```

```python
df['LGA'] = df['Cleaned_SA2_Name'].map(sa2_to_lga)
```

```python
missing_sa2 = df[~df['Cleaned_SA2_Name'].isin(sa2_to_lga.keys())]['Cleaned_SA2_Name'].unique()
print("SA2 names in df not found in the mapping:")
print(missing_sa2)
```

```python
df = df.merge(crime_df[['Rate per 100,000 population']], 
              left_on='LGA', 
              right_index=True, 
              how='left')
```

```python
# Rename the Rate per 100,000 population column to crime and drop the LGA column
df = df.rename(columns={'Rate per 100,000 population': 'crime'})
df = df.drop(columns=['LGA'])
```

```python
from sklearn.preprocessing import MinMaxScaler

# List of columns to consider
columns_to_consider = [
    'Population density (persons/km2)',
    'Health care and social assistance (no.)',
    'Education and training (no.)',
    'Public administration and safety (no.)',
    'Count of homeless persons (no.)',
    'Used at least one form of public transport (train, tram, bus, ferry) (no.)',
    'rental affordability'
]

# Check for missing values and impute with column average
for column in columns_to_consider:
    if df[column].isna().sum() > 0:
        print(f"Warning: Missing values in {column}. Imputing with column average.")
        df[column].fillna(df[column].mean(), inplace=True)

# Invert rental affordability
df['rental affordability'] = 1 / df['rental affordability']

# Normalize the features
scaler = MinMaxScaler()
df[columns_to_consider] = scaler.fit_transform(df[columns_to_consider])

# Compute the livability score
df['Livability Score'] = (
    0.1 * df['Population density (persons/km2)'] +
    0.2 * df['Health care and social assistance (no.)'] +
    0.2 * df['Education and training (no.)'] +
    0.2 * df['Public administration and safety (no.)'] -
    0.1 * df['Count of homeless persons (no.)'] +
    0.1 * df['Used at least one form of public transport (train, tram, bus, ferry) (no.)'] +
    0.1 * df['rental affordability']
)

# Get the top 10 livable suburbs
top_10_suburbs = df.nlargest(10, 'Livability Score')[['Cleaned_SA2_Name', 'Livability Score', 'rental affordability']]
top_10_suburbs
```

```python
# get the top 10 affodable suburbs
top_10_afford = df.nsmallest(10, 'rental affordability')[['Cleaned_SA2_Name', 'Livability Score', 'rental affordability']]
top_10_afford
```
