```python
import pandas as pd
import os
```

```python
#Find the SA2 area name for each row in the ABS data set

abs_2017 = pd.read_csv('../data/curated/2017ABS_csv')
for i in range(len(abs_2017['SA2 Area Name'])):
    abs_2017['SA2 Area Name'][i] = abs_2017['SA2 Area Name'][i].split('summary_ ', 1)[-1].split(' SA2', 1)[0].upper()
    
    if '-' in abs_2017['SA2 Area Name'][i]:

        abs_2017['SA2 Area Name'][i] = abs_2017['SA2 Area Name'][i].replace(' -', '')


abs_sa2 =[]
for i in range(len(abs_2017['SA2 Area Name'])):
    abs_sa2.append(abs_2017['SA2 Area Name'][i].upper())


#area codes to drop so we can get more rows filled
drop_words = ['NORTH', 'SOUTH', 'EAST', 'WEST', '(VIC.)']

#This will make Keilor East and Keilor West both just Keilor
for i in range(len(abs_2017['SA2 Area Name'])):
    for word in drop_words:
        if word in abs_2017['SA2 Area Name'][i]:
            word = ' '+word
            abs_2017['SA2 Area Name'][i] = abs_2017['SA2 Area Name'][i].replace(word, '')
```
