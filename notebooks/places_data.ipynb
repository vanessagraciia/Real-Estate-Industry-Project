```python
import requests
import pandas as pd

url = "https://data.melbourne.vic.gov.au/api/explore/v2.1/catalog/datasets/landmarks-and-places-of-interest-including-schools-theatres-health-services-spor/records?order_by=sub_theme"
limit = 100  # Maximum limit per request
offset = 0
all_records = []

while True:
    params = {"limit": limit, "offset": offset}
    response = requests.get(url, params=params)

    if response.status_code == 200:
        data = response.json()
        records = data['results']

        if not records:
            break  # No more records to fetchhs

        all_records.extend(records)
        offset += limit
    else:
        print("Request failed with status code:", response.status_code)
        break

# Create a DataFrame from all the records
df = pd.DataFrame(all_records)
```

```python
# Save the DataFrame to a CSV file
df.to_csv("../data/curated/melb_landmarks.csv", index=False)

print(f"Data saved to melbo_landmarks.csv with {len(all_records)} records.")
```
