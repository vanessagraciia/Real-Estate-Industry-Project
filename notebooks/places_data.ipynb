```python
import requests
import pandas as pd

#Accessing the API 
url = "https://data.melbourne.vic.gov.au/api/explore/v2.1/catalog/datasets/landmarks-and-places-of-interest-including-schools-theatres-health-services-spor/records?order_by=sub_theme"
limit = 100  #Maximum limit per request
offset = 0
all_records = []

#Iterating through to retrieve all the records until there are no more left
while True:
    params = {"limit": limit, "offset": offset}
    
    #Sending a GET request for the API using specified parameters
    response = requests.get(url, params=params)

    if response.status_code == 200:
        data = response.json()
        # Parse the JSON data from the response
        records = data['results']

        #Check if there are any records left to fetch, if no more then stop.
        if not records:
            break  

        all_records.extend(records)
        offset += limit
    else:
        print("Request failed with status code:", response.status_code)
        break

#Create a DataFrame for all the records
df = pd.DataFrame(all_records)
df
```

```python
df.to_csv("../data/landing/melb_landmarks.csv", index=False)
print(f"Data saved to melbo_landmarks.csv with {len(all_records)} records.")
```
